```python
# import plotly.graph_objects as go
# import plotly.express as px
# from jupyter_dash import JupyterDash
# import dash_core_components as dcc
# import dash_html_components as html
# from dash.dependencies import Dash, Input, Output
from dash import Dash, dcc, html, Input, Output
import plotly.express as px
```

```python
fig = px.bar(df1, x ='sgpartido', y = 'vlrliquido', color = 'sguf', labels={'vlrliquido':'Valor Gasto','sgpartido':'Partido', 'sguf':'Estado'}, height=400)
fig.show()
```

```python
fig2 = px.pie(df1, values='vlrliquido', names='sgpartido', title='Gasto por Partido')
fig2.show()
```
